```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numbers= np.random.rand(200)
```

```python
sns.distplot(numbers)
```

```python
sns.distplot(numbers, color='red')
```

```python
 label_dist= pd.Series(numbers, name='var_x')
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist, vertical='True')
```

```python
sns.distplot(label_dist, vertical='True', color='green')
```

```python
sns.distplot(label_dist, rug='False', hist=False,rug_kws={"color": "y","lw": 10})
```
