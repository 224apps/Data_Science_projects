```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=tips['total_bill'])
```

```python
ax= sns.boxplot(x='day', y='total_bill',data=tips)
```

```python
ax= sns.boxplot(x='day', y='total_bill',data=tips, hue='smoker')
```

```python
ax= sns.boxplot(x='day', y='total_bill',data=tips, palette='Set2')
```

```python
ax= sns.boxplot(x='day', y='total_bill',data=tips, linewidth= 2.5)
```

```python
ax = sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```
